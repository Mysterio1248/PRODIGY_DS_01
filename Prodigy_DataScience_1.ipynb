```python
import pandas as pd
import matplotlib.pyplot as plt

data_url = "/content/TaskDataset1.csv"
df = pd.read_csv(data_url)

gender_employment_source_counts = df.groupby(['Sex', 'EmploymentStatus', 'RecruitmentSource']).size().unstack()

fig, ax = plt.subplots(figsize=(10,6))
gender_employment_source_counts.plot(kind='bar', stacked=True, ax=ax)

ax.set_xlabel('Gender and Employment Status')
ax.set_ylabel('Count')
ax.set_title('Distribution of Employment Status and Recruitment Source by Gender')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
